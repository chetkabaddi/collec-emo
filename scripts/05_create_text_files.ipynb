```python
## helper function to return the name of the output file

def output_file_name_maker(input_file_name, desc, dev, region, extension) :
    year = '2020'
    file_month = input_file_name.split('_')[0].lower()
    if file_month == 'jan':
        month = '01'
    elif file_month == 'feb':
        month = '02'
    elif file_month == 'mar':
        month = '03'
    elif file_month == 'apr':
        month = '04'
    elif file_month == 'may':
        month = '05'
    date = input_file_name.split('_')[1]
    full_date = year+month+date
    output_file_name = desc + '_'+region+ '_'+ full_date+'_'+dev+'.'+extension
    return output_file_name
```
