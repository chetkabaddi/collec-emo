```python
region_list = ['usa','ind','uk', 'sgp', 'arg','aus','austria','can', 'chile', 'colombia','costa', 'fra', 'ger', 'ita', 'netherlands', 'nz', 'rus', 'esp', 'swiss', 'ven', 'bra', 'mex', 'per', 'tza', 'zaf', 'eth','nga']

for f in region_list:
    os.mkdir("../datewise_countrywise_old_1/"+f)
```

```python
scenario_one_old_keywords = ["coronavirus", "2019nCov", "corona virus", "wuhan", "pneumonia", "pneumonie", "neumonia" ,"Lungenentzündung", "covid19"]
scenario_one_new_keywords = ["COVID19", "CoronavirusPandemic", "COVID-19", "CoronaOutbreak", "WuhanVirus", "wuhan", "pneumonia", "pneumonie", "neumonia" ,"Lungenentzündung", "covid19"]
scenario_two_old_keywords = ["coronavirus", "2019nCov", "corona virus"] #, "wuhan", "pneumonia", "pneumonie", "neumonia" ,"Lungenentzündung", "covid19"]
scenario_two_new_keywords = ["COVID19", "CoronavirusPandemic", "COVID-19", "CoronaOutbreak", "WuhanVirus"] #, "wuhan", "pneumonia", "pneumonie", "neumonia" ,"Lungenentzündung", "covid19"]
```

```python
def s_one(file_path, dir_old, dir_new) :
    df = pd.read_csv(file_path, lineterminator='\n')
    df = df.drop('Unnamed: 0' , axis=1)
    x_idx = set()
    y_idx = set()
    f_name = file_path.split('/')[-1]
    for idx, row in df.iterrows() :
        if(re.compile('|'.join(scenario_one_old_keywords),re.IGNORECASE).search(row.full_text)):
            x_idx.add(idx)
    for idx, row in df.iterrows() :
        if(re.compile('|'.join(scenario_one_new_keywords),re.IGNORECASE).search(row.full_text)):
            y_idx.add(idx)
    filtered_idx = x_idx.union(y_idx)
    tweets_old = df.iloc[list(x_idx)]
    tweets_new = df.iloc[list(y_idx)]
    print(len(tweets_old))
    print(len(tweets_new))
#     tweets_unclassified = df.drop(filtered_idx)
#     print(len(tweets_unclassified))
#     tweets_old = pd.concat([tweets_old,tweets_unclassified])
#     tweets_new = pd.concat([tweets_new,tweets_unclassified])
#     print(len(tweets_old))
#     print(len(tweets_new))
    tweets_old.to_csv(dir_old+f_name)
    tweets_new.to_csv(dir_new+f_name)
    return tweets_old , tweets_new

def s_two(file_path, dir_old, dir_new) :
    df = pd.read_csv(file_path, lineterminator='\n')
    df = df.drop('Unnamed: 0' , axis=1)
    x_idx = set()
    y_idx = set()
    f_name = file_path.split('/')[-1]
    for idx, row in df.iterrows() :
        if(re.compile('|'.join(scenario_two_old_keywords),re.IGNORECASE).search(row.full_text)):
            x_idx.add(idx)
    for idx, row in df.iterrows() :
        if(re.compile('|'.join(scenario_two_new_keywords),re.IGNORECASE).search(row.full_text)):
            y_idx.add(idx)
    filtered_idx = x_idx.union(y_idx)
    tweets_old = df.iloc[list(x_idx)]
    tweets_new = df.iloc[list(y_idx)]
    print(len(tweets_old))
    print(len(tweets_new))
#     tweets_unclassified = df.drop(filtered_idx)
#     print(len(tweets_unclassified))
#     tweets_old = pd.concat([tweets_old,tweets_unclassified])
#     tweets_new = pd.concat([tweets_new,tweets_unclassified])
#     print(len(tweets_old))
#     print(len(tweets_new))
    tweets_old.to_csv(dir_old+f_name)
    tweets_new.to_csv(dir_new+f_name)
```

```python
def filter_tweets(input_dir) :
    
    t = time.localtime()
    current_time = time.strftime("%H:%M:%S", t)
    print(current_time)
    print("country : {}".format(input_dir.split('/')[-1]))
    op_dir_old ="../datewise_countrywise_old_1/"+input_dir.split('/')[2]+'/' 
    op_dir_new ="../datewise_countrywise_new_1/"+input_dir.split('/')[2]+'/' 
#     print(op_dir_old)
#     print(op_dir_new)
    for _,_,files in os.walk(input_dir) :
        print(files)
    for f in files :
        date = int(f.split('_')[1])
        month = f.split('_')[0]
#         print(f)
        file_name = input_dir+'/'+f
        print(file_name)
        if(month in ['Jan', 'Feb', 'Mar']) :
            if(month == 'Jan' and date<27) :
                print("continue")
                continue
            elif(month =="Mar" and date >=28 ):
                print("scenario two")
                s_two(file_name, op_dir_old, op_dir_new)
            else :
                print("scenario one")
                s_one(file_name, op_dir_old, op_dir_new)
        else :
            print("scenario two")
            s_two(file_name, op_dir_old, op_dir_new)
```

```python
for r in region_list :
    path = "../datewise_countrywise/"+r
    filter_tweets(path)
```
