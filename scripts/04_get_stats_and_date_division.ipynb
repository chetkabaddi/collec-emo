```python
## Function to calculate the total word count on each day for a particular country
def get_stats(df) :
    word_count = 0
    for i in df['full_text'] :
        word_count+= len(re.findall(r'\w+', i))
    return word_count

## Function to get datewise tweets for a particular country
def get_datewise(df,date,month,country,suffix,outputF) :
    temp = df[(df['month']==month) & (df['date']==int(date))]
    tweet_count = 0
    total_words = 0
    avg_words_per_tweet = 0 
    if(len(temp)>0):
        print("tweets in "+country+ " = " + str(len(temp)))
        tweet_count = len(temp)
#         avg_words_per_tweet = (get_stats(temp)/tweet_count)
        total_words = get_stats(temp)
        output_file =outputF+".csv"
        temp.to_csv(output_file)
    
    return tweet_count , total_words
```

```python
region = 'bra'
# german : de spanish : es french:fr italian :it dutch:nl russian:ru
lang = 'es'
country = 'brazil'
```

```python
## getting list of files for a particular country

flag = 0
completed_tweets_path = '../datewise_countrywise/'+region+'/'
for _,_,files in os.walk(completed_tweets_path) :
    print(files)
print(len(files))
total_tweets = 0

## Traversing through the files and creating a dataframe by concatening them.

for file in files :
    if(flag==0) :
        df=pd.read_csv(completed_tweets_path+file, index_col=False)
        flag=1
        print(len(df))
        total_tweets+=len(df)
    else :
        df_temp = pd.read_csv(completed_tweets_path+file , index_col=False, error_bad_lines=False)
        print(len(df_temp))
        df = pd.concat([df,df_temp])
        total_tweets+=len(df_temp)
```

```python
df = df.reset_index(drop = True)
df
```

```python
df.drop_duplicates(subset='tweet_id', keep='first', inplace=True)
```

```python
df = df[df['language']==lang]
df
```

```python
## Dictionaries to help when sorting by date

months = {'Jan':'01' ,'Feb':'02' , 'Mar':'03', 'Apr':'04', 'May':'05'}
days = {
    'Jan' : 31,
    'Feb' : 29,
    'Mar' : 31,
    'Apr' : 30,
    'May' : 18,
}
```

```python
tweets = []
wpt_stat = []
mth = []
dt = []
tweet_sum = 0

for m in months :
    for i in range(1,days[m]+1) :
        print(m)
        print(i)
        if(i <=9) :
            date = '0'+str(i)
        else :
            date= str(i)
        
        mth.append(m)
        dt.append(i)
        
        outputF = "../datewise_countrywise/"+region+"/"+"datewise_"+region+"_2020"+months[m]+date + "_HS"
        tot_tweet , wpt = get_datewise(df,date,m,country,region,outputF)
        tweet_sum+=tot_tweet
        tweets.append(tot_tweet)
        wpt_stat.append(wpt)
```

```python
## storing the list of different stats into a dictionary; easy to convert to a dataframe

stats_arg = {
    'date': dt,
    'month': mth,
    'tweets_arg' :tweets,
    'wpt_arg' : wpt_stat
}
```

```python
stats_arg = pd.DataFrame(stats_arg)
```
