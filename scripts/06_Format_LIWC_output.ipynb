```python
df = pd.DataFrame({'Data': [10, 20, 30, 20, 15, 30, 45]})

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('pandas_con.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Sheet1']

# Apply a conditional format to the cell range.
worksheet.conditional_format('B2:B3', {'type': '3_color_scale'})

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

## Final highlight script

```python
df = pd.read_csv("../LIWC outputs/LIWC2015 Results (zaf (116 files)).csv" )
df
```

```python
df_filename = df['Filename']
df_liwc = df.drop('Filename', axis=1)
df_liwc
```

```python
date = []
for name in df_filename :
    date_segment = name.split('_')[2]
#     print(date_segment)
    year = date_segment[:4]
    month = date_segment[4:6]
    day = date_segment[6:]
#     print(year+" "+month+" "+day)
    date_string = year+'-'+month+'-'+day
#     print(date_string)
    date_object = pd.to_datetime(date_string).date()
    print(date_object)
    date.append(date_object)
```

```python
date = pd.Series(date, name='date')
date
```

```python
df_new = pd.concat([df_filename, date], axis=1)
```

```python
df_new
```

```python
df_final = pd.concat([df_new, df_liwc], axis=1)
df_final
```

```python
idx_list = []
for idx, row in df_final.iterrows() :
    if row['WC'] < 5000 :
        idx_list.append(idx)
```

```python
idx_list
```

```python
#df = pd.read_csv("../LIWC2015 Results (arg (119 files)).csv")
writer = pd.ExcelWriter('LIWC output highlighted_zaf_HS.xlsx', engine='xlsxwriter')
df_final.to_excel(writer, sheet_name='Sheet1')
workbook  = writer.book
worksheet = writer.sheets['Sheet1']
format1 = workbook.add_format({'bg_color': '#FFC7CE'})
for i in idx_list :
    row_range = 'A'+str(i+2)+':'+'CZ'+str(i+2)
    worksheet.conditional_format(row_range, {'type':'3_color_scale'})
writer.save()
```
